```
setosa_x = iris.data[:,0][iris.target ==0]
setosa_y = iris.data[:,1][iris.target ==0]
versicolor_x = iris.data[:,0][iris.target==1]
versicolor_y = iris.data[:,1][iris.target==1]
X = np.concatenate((setosa_x , versicolor_x))
Y = np.concatenate((setosa_y , versicolor_y))
targets = iris['target'][:100]
def colorit(data, a, b):
  colors = np.copy(data)
  for i in range (100):
    if Y[i] > a*X[i] +b:
      colors[i] = 1 - colors[i]
  return colors
def find():
  max = [0,0,0]
  for i in np.arange(-10,10, 0.1):
    for j in np.arange(-10,10,0.1):
      colors = colorit(targets, i, j)
      if np.sum(colors) > max[0]:
        max[0] = np.sum(colors)
        max[1] = i
        max[2] = j
  return max[1], max[2]

a, b = find()
print(a, b)
colors = colorit(targets, a, b)
print(colors)


plt.title('Iris')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
kolory = ListedColormap(['red', 'green'])
plt.scatter(X, Y, c=colors, marker='o',cmap=kolory, linewidth=0, s=15)
prosta = np.arange(4.25, 8)
plt.plot(prosta, a*prosta + b)
plt.xticks(range(3,10))
plt.yticks(range(1,6))
plt.ylim(1.5,5)
plt.xlim(4,7.2)
plt.show()

```
