a) Wczytaj dane IRIS. Można to szybko zrobić korzystając z biblioteki SciKit za pomocą polecenia: 

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from matplotlib.colors import ListedColormap
iris = datasets.load_iris() 
```

Zobacz co zawiera iris.data, iris.target, iris.feature_names, iris.target_names. 


```
#print(iris.data)
print(np.sum(iris.data[iris.target == 0], axis=0))
#print(iris.data[iris.target == 0])
def mean(X):
  return np.sum(X, axis=0) / np.size(X, axis=0)
print(type(iris.target))
print(mean(iris.data[iris.target == 0]))
```

b) Przedstaw wczytane dane na wykresie w poniższy sposób (zwróć uwagę na kolory i typ
znaczników, opisy osi i tytuł wykresu): 

```
X = iris.data[:, 0]
Y = iris.data[:, 1]

plt.title('Iris')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
kolory = ListedColormap(['red', 'blue', 'yellow'])
plt.scatter(X, Y, c=iris['target'], marker='o', linewidth=0, cmap=kolory, s=15)
plt.xticks(range(3,10))
plt.yticks(range(1,6))
plt.show()
plt.savefig('iris_plot.png')
```
