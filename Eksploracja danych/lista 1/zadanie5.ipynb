```
a = 2
b = 5
X = np.arange(1,11)
Y = np.array([10, 10, 11, 12, 18, 18, 19, 26, 19, 26])
plt.plot(X, Y, 'r.')
plt.plot(X, a*X + b)
def dist(x, y, a, c):
  return np.abs(a*x + -1*y + c)/np.sqrt(a**2 + 1)

sumofdists = sum(dist(X, Y, a, b))
print(sumofdists)
for i in np.arange(-10,10,0.1):
  for j in np.arange(-10,10,0.1):
      if sum(dist(X, Y, i, j)) < sumofdists:
        sumofdists = sum(dist(X, Y, i, j))
        a = i
        b = j
plt.plot(X, a*X + b)
print(sumofdists)
```
