```
import numpy as np 
import math
a = np.arange(1, 101)
print(a)
b = np.arange(1, 100, 2)
print(b)

c = np.arange(0, 2.01, 0.01)
c = c - 1
c = c * np.pi
print(c)
d = np.delete(c, 100)
print(d)

e = np.zeros((1, 100))
for i in range (1,101):
  sin = math.sin(i)
  if sin>0:
    e[0][i] = sin
print(e)

A = np.reshape(a, (10, 10))
print(A)

B = np.zeros((100,100))
B[0][0] = 1
B[0][1] = 99
B[99][99] = 100
B[99][98] = 1
for i in range(1, 99):
  for j in range (i-1, i+2):
    if i == j:
      B[i][j] = i+1
    elif i>j:
      B[i][j] = 100-i
    else:
      B[i][j] = 99-i
print(B)

C = np.ones((10,10))
for i in range(10):
  for j in range(0, i):
    C[i][j] = 0
print(C)

triangl = np.arange(1, 101)
fact = np.ones(100)
  
for i in range (1, 100):
  triangl[i] = triangl[i-1] + i+1
  fact[i] = fact[i-1] * (i+1)
D = np.reshape((np.concatenate((triangl, fact), axis=0)), (2,100))
print(D)

E = np.zeros((100, 100))
for i in range(100):
  for j in range(i, 100):
    if (j+1)%(i+1) == 0:
      E[i][j] = 1
print(E)
```
