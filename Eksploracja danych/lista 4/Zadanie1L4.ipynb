```
% matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.cluster import KMeans, MiniBatchKMeans, Birch, DBSCAN
```

```
import matplotlib.colors as colors
from itertools import cycle

def plotClustering(X, plt_labels):
    plt_colors = cycle(['b', 'g', 'r', 'c', 'm', 'y', 'k'])

    plt_K = np.unique(plt_labels).size

    for k in range(-1, plt_K):
        color = next(plt_colors)
        mask = (plt_labels == k)
        plt.plot(X[mask, 0], X[mask, 1], 'w', markerfacecolor=color, marker='o')
    
    plt.show()
```

```
from sklearn import datasets

centers_ = [[1, 1], [3, 3], [5, 1]]
X, labels = datasets.make_blobs(n_samples=3000, n_features=2, centers=centers_, cluster_std=0.5)
```

```
birch = Birch(threshold=0.2, n_clusters=6)
birch.fit(X)
plt_K = np.unique(birch.labels_).size
plt_colors = cycle(['b', 'g', 'r', 'c', 'm', 'y'])
for k in range(plt_K):
  color = next(plt_colors)
  mask = (birch.labels_ == k)
  plt.scatter(X[mask, 0], X[mask, 1],edgecolors='w',c=color)

plt.scatter(birch.root_.centroids_[:,0], birch.root_.centroids_[:,1],marker='o', edgecolors='w', color='k', s=70)
morecenters=birch.subcluster_centers_
plt.scatter(morecenters[:,0], morecenters[:,1],marker='o', edgecolors='w', color='orange', s=35)
```
