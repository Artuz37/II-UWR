```
import numpy as np
import matplotlib.pyplot as plt
import os
import time
from IPython.display import clear_output

def play(width = 2, height = 2, steps=100):
    BOARD = np.random.choice([0, 1], width * height)
    BOARD = BOARD.reshape(height, width)

    dx = [1, -1, 0, 0, 1, 1, -1, -1]
    dy = [0, 0, 1, -1, 1, -1, 1, -1]

    def print_board():
        clear_output()        
        plt.imshow(BOARD, animated=True)
        plt.show()
        time.sleep(0.5)

    for no in range(steps):
        for i in range(height):
            for j in range(width):
                alive_n = 0
                for x in range(len(dx)):
                    new_x = i + dx[x]
                    new_y = j + dy[x]

                    if 0 <= new_x < height \
                        and 0 <= new_y < width \
                            and BOARD[new_x,new_y] == 1:
                        alive_n += 1


                if alive_n == 3 and BOARD[i,j] == 0:
                    BOARD[i,j] = 1

                if BOARD[i,j] == 1:
                    if alive_n != 2 and alive_n != 3:
                        BOARD[i,j] = 0
        print_board()
        





play(10, 10, 100)
```
