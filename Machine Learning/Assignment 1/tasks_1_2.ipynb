g) Declare ismagic(A) function which checks if matrix $A$ is a magic square and returns a boolean.

```
def ismagic(A):
    x = np.apply_along_axis(sum, axis=1, arr=A)
    s = x[0]
    y = np.apply_along_axis(sum, axis=0, arr=A)
    return np.all(x==s) and np.all(y==s) and sum(np.diagonal(A))==s and sum(np.diagonal(np.fliplr(A)))==s

print(ismagic(np.array([[1,1], [2,2]])))
print(ismagic(np.array([[2,7,6],[9,5,1],[4,3,8]])))
```
